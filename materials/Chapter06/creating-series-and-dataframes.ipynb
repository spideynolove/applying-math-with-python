```python
import pandas as pd
import numpy as np
```

```python
from numpy.random import default_rng
rng = default_rng(12345)
```

```python
diff_data = rng.normal(0, 1, size=100)
cumulative = diff_data.cumsum() 
```

```python
data_series = pd.Series(diff_data)
print(data_series)
```

```python
data_frame = pd.DataFrame({
    "diffs": data_series, 
    "cumulative": cumulative 
})
```

```python
print(data_frame)
```
