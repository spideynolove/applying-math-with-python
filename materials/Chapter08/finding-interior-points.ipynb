```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from shapely.geometry import Polygon, Point
```

```python
polygon = Polygon(
    [(0, 2), (-1, 1), (-0.5, -1), (0.5, -1), (1, 1)],
)
```

```python
fig, ax = plt.subplots()
poly_patch = mpl.patches.Polygon(polygon.exterior.coords, ec=(0,0,0,1), fc=(0.5,0.5,0.5,0.4))
ax.add_patch(poly_patch)
ax.set(xlim=(-1.05, 1.05), ylim=(-1.05, 2.05))
ax.set_axis_off()
```

```python
p1 = Point(0.0, 0.0)
p2 = Point(-1.0, -0.75)
```

```python
ax.plot(0.0, 0.0, "k*")
ax.annotate("p1", (0.0, 0.0), (0.05, 0.0))
ax.plot(-0.8, -0.75, "k*")
ax.annotate("p2", (-0.8, -0.75), (-0.8 + 0.05, -0.75))
```

```python
plt.show()
```

```python
print("p1 inside polygon?", polygon.contains(p1))
print("p2 inside polygon?", polygon.contains(p2))
```
