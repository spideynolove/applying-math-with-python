```python
import numpy as np
```

```python
from numpy import linalg
```

```python
A = np.array([[3, -1, 4], [-1, 0, -1], [4, -1, 2]])
```

```python
v, B = linalg.eig(A)
```

```python
i = 0  
lambda0 = v[i]
lambda0
```

```python
x0 = B[:, i]  
x0
```

```python
linalg.norm(x0)
```

```python
lhs = A @ x0
rhs = lambda0*x0
linalg.norm(lhs - rhs)
```
