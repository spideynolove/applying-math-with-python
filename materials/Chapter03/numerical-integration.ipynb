```python
import numpy as np
```

```python
def erf_integrand(t):
    return np.exp(-t**2)
```

```python
from scipy import integrate
```

```python
val_quad, err_quad = integrate.quad(erf_integrand, -1.0, 1.0)
val_quad, err_quad
```

```python
val_quadr, err_quadr = integrate.quadrature(erf_integrand, -1.0, 1.0)
val_quadr, err_quadr
```
