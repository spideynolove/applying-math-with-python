```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
from shapely.geometry import Polygon
from shapely.ops import triangulate
```

```python
polygon = Polygon(
    [(2.0, 1.0), (2.0, 1.5), (-4.0, 1.5), (-4.0, 0.5), (-3.0, -1.5),
     (0.0, -1.5), (1.0, -2.0), (1.0, -0.5), (0.0, -1.0), (-0.5, -1.0),
     (-0.5, 1.0)],
    holes=[np.array([[-1.5, -0.5], [-1.5, 0.5], [-2.5, 0.5], [-2.5, -0.5]])]
)
```

```python
fig, ax = plt.subplots()
plt_poly = mpl.patches.Polygon(polygon.exterior.coords, ec=(0,0,0,1), fc=(0.5,0.5,0.5,0.4), zorder=0)
ax.add_patch(plt_poly)
plt_hole = mpl.patches.Polygon(polygon.interiors[0].coords, ec="k", fc="w")
ax.add_patch(plt_hole)
ax.set(xlim=(-4.05, 2.05), ylim=(-2.05, 1.55))
ax.set_axis_off()
```

```python
triangles = triangulate(polygon)
```

```python
filtered = filter(lambda p: polygon.contains(p), triangles)
```

```python
patches = map(lambda p: mpl.patches.Polygon(p.exterior.coords), filtered)
col = mpl.collections.PatchCollection(patches, fc="none", ec="k")
```

```python
ax.add_collection(col)
```

```python
plt.show()
```
