```python
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.feature import canny
```

```python
image = imread("mandelbrot.png", as_gray=True)
```

```python
edges = canny(image, sigma=0.5)
```

```python
fig, ax = plt.subplots()
ax.imshow(edges, cmap="gray_r")
ax.set_axis_off()
```

```python
plt.show()
```
