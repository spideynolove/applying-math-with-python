```python
import dask.dataframe as dd
```

```python
data = dd.read_csv("sample.csv", dtype={"number": "object"})
print(data.head())
```

```python
sum_data = data.lower + data.upper
print(sum_data)
```

```python
result = sum_data.compute()
print(result.head())
```

```python
means = data[["lower", "upper"]].mean().compute()
print(means)
```
