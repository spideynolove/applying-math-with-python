```python
from scipy import optimize
from math import exp
```

```python
%run ../functions.py
```

```python
def f(x):
    return x*(x - 2)*exp(3 - x)
```

```python
def fp(x):
    return -(x**2 - 4*x + 2)*exp(3 - x)
```

```python
# Using x1 = 1.5 and the secant method
# optimize.newton(f, 1., x1=1.5) 
optimize.newton(f, 1., x1=2.1)
```

```python
# Using Newton-Raphson method
optimize.newton(f, 1., fprime=fp) 
# optimize.newton(f, 2.5, fprime=fp) 
```
