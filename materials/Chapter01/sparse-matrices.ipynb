```python
import numpy as np
from scipy import sparse
from scipy.sparse import linalg
```

```python
T = sparse.diags([-1, 2, -1], (-1, 0, 1), shape=(5, 5), format="csr")   # row based
```

```python
type(T)
```

```python
temp = T.toarray()
type(temp)
```

```python
temp
```

```python
linalg.spsolve(T.tocsr(), np.array([1, 2, 3, 4, 5]))
```

```python
T2 = sparse.diags([-1, 2, -1], (-1, 0, 1), shape=(5, 5), format="csc")  # column based
T2.toarray()
```

```python
type(T2)
```

```python
linalg.spsolve(T2.tocsc(), np.array([1, 2, 3, 4, 5]))
```
