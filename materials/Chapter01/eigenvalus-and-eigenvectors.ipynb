```python
import numpy as np
```

```python
A = np.array([[3, -1, 4], [-1, 0, -1], [4, -1, 2]])
A
```

```python
# linear algebra, Compute the eigenvalues and right eigenvectors of a square array
v, B = np.linalg.eig(A)
v, B
```

```python
# Extract a diagonal or construct a diagonal array.
D = np.diag(v)
D
```

```python
A_reconstructed = np.dot(B, np.dot(D, np.linalg.inv(B)))
A_reconstructed
```
